Here we are importing the '**mnist_784**' dataset from the sklearn and then scaling the pixel values to be in [0,1]

```python
from sklearn.datasets import fetch_openml

X, Y = fetch_openml('mnist_784', version=1, return_X_y=True, as_frame=False,parser='auto')
X = X /255
```

Mnist dataset originally has 60000 instances in training dataset and 10000 in test dataset. Here we are splitting the data in train and test data.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=10000, random_state=42)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

Plot few instances from the training dataset to make sure it looks like handwritten digits

```python
from matplotlib import pyplot as plt
for i in range(3):
 plt.matshow(X_train[i].reshape(28,28))
```

Plot few instances from the test dataset to make sure it looks like handwritten digits

```python
for i in range(3):
 plt.matshow(X_test[i].reshape(28,28))
```

MLP classifier with logistic as activation function and 50 hidden layers

```python
%%time
from sklearn.neural_network import MLPClassifier

model1 = MLPClassifier(hidden_layer_sizes=(50,), activation = 'logistic', verbose=True)

model1.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score

Y_pred = model1.predict(X_test)
model1_accuracy = accuracy_score(Y_test, Y_pred)
print('Accuracy on test set:', model1_accuracy)
```

MLP classifier with logistic as activation function and 100 hidden layers

```python
%%time
from sklearn.neural_network import MLPClassifier

model2 = MLPClassifier(hidden_layer_sizes=(100,), activation = 'logistic',verbose=True)
model2.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score

Y_pred = model2.predict(X_test)
model2_accuracy = accuracy_score(Y_test, Y_pred)
print('Accuracy on test set:', model2_accuracy)
```

MLP classifier with relu activation function and 100 hidden layers

```python
%%time
from sklearn.neural_network import MLPClassifier

model3 = MLPClassifier(hidden_layer_sizes=(100,), activation = 'relu', verbose=True)
model3.fit(X_train, Y_train)
```

```python

Y_pred = model3.predict(X_test)
model3_accuracy = accuracy_score(Y_test, Y_pred)
print('Accuracy on test set:', model3_accuracy)
```

There is not much difference in accuracy with the change of activation function but there is a change of around 1% in accuracy with increasing the number of layers to 100.

SVM


```python
%%time
from sklearn.svm import SVC
model4 = SVC(C=3)
model4.fit(X_train,Y_train)
Y_pred = model4.predict(X_test)
model4_accuracy = accuracy_score(Y_test,Y_pred)
print(model4_accuracy)
```

SVM method has better accuracy but the training time is much longer than the MLP Classifer.

CNN

```python
%%time
from keras import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.utils import to_categorical

X_CNN_train = X_train
X_CNN_test = X_test
print(X_CNN_train.size)
print(X_CNN_test.size)



X_CNN_train = X_CNN_train.reshape(60000,28, 28, 1)
X_CNN_test = X_CNN_test.reshape(10000,28, 28, 1)

Y_CNN_train = Y_train
Y_CNN_test = Y_test

Y_CNN_train = to_categorical(Y_CNN_train)
Y_CNN_test = to_categorical(Y_CNN_test)



model5 = Sequential()
model5.add(Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=(28,28,1)))
model5.add(Conv2D(filters=32, kernel_size=3, activation='relu'))
model5.add(MaxPooling2D(pool_size=(2, 2)))
model5.add(Flatten())


model5.add(Dense(128, activation = 'relu'))
model5.add(Dense(10, activation = 'softmax'))

from keras.backend import categorical_crossentropy

model5.compile(optimizer='adam', loss=categorical_crossentropy, metrics=['accuracy'])
model5.fit(X_CNN_train, Y_CNN_train, batch_size=128, epochs=10, validation_split=0.1)

score = model5.evaluate(X_CNN_test, Y_CNN_test, verbose=True)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```

30 degree rotation

```python

import numpy as np
from numpy import random
X = X.reshape(70000,28,28)
def generateNewImages(X, Y, size, angle):
  X_rotated = np.zeros(shape = (size, 28,28))
  Y_rotated = np.zeros(shape = (size,),dtype='object')

  for k in range (size):
    rnd = random.randint(size)
    if X_rotated[k] not in X_rotated:
      X_rotated[k]= rotate(X_train[rnd], angle, reshape=False)
      Y_rotated[k]=Y[rnd]
    else:
      rnd = random.randint(size)
      X_rotated[k]= rotate(X[rnd], angle, reshape=False)
      Y_rotated[k]=Y[rnd]


  return X_rotated, Y_rotated

size = int(0.3 * X.shape[0])
X_rotated, Y_rotated = generateNewImages(X, Y, size, 30)
```

```python
print(X_rotated.shape)
print(Y_rotated.shape)

X_rotated_train, X_rotated_test, Y_rotated_train, Y_rotated_test = train_test_split(X_rotated, Y_rotated, test_size=3000, random_state=42)

print(X_rotated_train.shape)
print(X_rotated_test.shape)
print(Y_rotated_train.shape)
print(Y_rotated_test.shape)
```
